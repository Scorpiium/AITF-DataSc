```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/content/sample_data/california_housing_test.csv')
df.head()
```

`df.info()`: Ini memberikan ringkasan informasi DataFrame, termasuk tipe data setiap kolom dan jumlah nilai non-null. Mengidentifikasi kolom mana yang mungkin memerlukan penanganan nilai yang hilang atau konversi tipe data.

```python
df.info()
```

`df.shape`: Perintah ini digunakan untuk melihat dimensi dataset.

```python
df.shape
```

`df.isnull().sum()`: Memeriksa jumlah nilai yang hilang di setiap kolom. Mengetahui keberadaan dan jumlah nilai yang hilang sangat penting karena dapat memengaruhi analisis.

```python
df.isnull().sum()
```

`df.describe()`: Menghasilkan statistik deskriptif untuk kolom-kolom numerik dalam dataset (jumlah, rata-rata, standar deviasi, nilai minimum, nilai maksimum, kuartil).

```python
df.describe()
```

`df.corr()`: Ini menghitung korelasi antar kolom numerik dalam DataFrame. Nilai korelasi berkisar antara -1 hingga 1, menunjukkan kekuatan dan arah hubungan linier antara dua variabel.

```python
df.corr()
```

`sns.heatmap(df.corr(), annot=True, cmap='coolwarm')`: Setelah menghitung matriks korelasi dengan df.corr(), heatmap divisualisasikan menggunakan seaborn (sns.heatmap). Heatmap adalah cara yang sangat efektif untuk memvisualisasikan matriks korelasi.

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```
