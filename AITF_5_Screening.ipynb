Delimiter digunakan untuk memisahkan database tergantung karakter yang digunakan dalam argumen, dalam contoh menggunakan semi-colon.

```python
df = pd.read_csv('Data_Nasabah.csv', delimiter=';')
df.head()
```

```python
# Jumlah data (baris dan kolom)
print("Jumlah data (baris, kolom):", df.shape)

# Informasi tentang tipe data dan skema pengkodean
print("\nInformasi tipe data dan skema pengkodean:")
print(df.info())

# Statistik deskriptif untuk data numerik
print("\nStatistik deskriptif untuk data numerik:")
print(df.describe())

# Melihat beberapa baris pertama data
print("\nBeberapa baris pertama data:")
print(df.head())

# Menampilkan jumlah nilai unik untuk setiap kolom (untuk memahami kategori)
print("\nJumlah nilai unik untuk setiap kolom:")
for column in df.columns:
    print(f"{column}: {df[column].nunique()}")
```

##### Enconding

```python
# Check unique values for non-numerical columns
print("Unique values for 'jenis_kelamin':", df['jenis_kelamin'].unique())
print("Unique values for 'jenis_produk':", df['jenis_produk'].unique())
print("Unique values for 'pengguna_mobile_banking':", df['pengguna_mobile_banking'].unique())
```

```python
df['jenis_kelamin'] = df.jenis_kelamin.map({'Laki-Laki': 0, 'Perempuan': 1})
df['jenis_produk'] = df.jenis_produk.map({'tabungan': 0, 'kartu_kredit': 1, 'deposito': 2})
df['pengguna_mobile_banking'] = df.pengguna_mobile_banking.map({'YA': 1, 'TIDAK': 0})

df.head()
```
